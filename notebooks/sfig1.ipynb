# Fig 1S

Basic simulations with two neighbors over 1D (with 30 cells) and 2D (with 100 cells) spatial grids considering 2-4 tasks. 

- n - number of cells (or locations)
- m - number of tasks
- d - number of dimensions
- n_neigh - number of neighbors to consider


```python
n_neigh = 2
n_2d = 100

kwargs_dict = {
               '1d2t': {'n': 30, 'm': 2, 'd':1, 'n_neigh':n_neigh}, 
               '1d3t': {'n': 30, 'm': 3, 'd':1, 'n_neigh':n_neigh}, 
               '1d4t': {'n': 30, 'm': 4, 'd':1, 'n_neigh':n_neigh, 'test': False}, 

               '2d2t': {'n': 100, 'm': 2, 'd':2, 'n_neigh':n_neigh}, 
               '2d3t': {'n': 100, 'm': 3, 'd':2, 'n_neigh':n_neigh}, 
               '2d4t': {'n': 100, 'm': 4, 'd':2, 'n_neigh':n_neigh, 'test': False}, 
            }
```

```python
for desc, kwargs in kwargs_dict.items():
    
    print(desc)
    G, T, L, obj = P.po.solve_opt(**kwargs)
    pl = P.pl.plot_sol(G, T, L, kwargs['n_neigh'], obj, tit=None, paper=True, remove_labels=True)
    
    if save:
        pl.save(os.path.join(outdir, 's1_' + desc + '.png'), scale_factor=5)
    pl.display()
```
