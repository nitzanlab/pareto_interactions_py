```python
import os
import random
import numpy as np

import pareto_interactions_py as P
from scipy.spatial.distance import squareform, pdist

random.seed(2)

save = False
outdir = '../output/'

desc_to_run = [ 
    'regular',
    'exp_inhibit',
    'stimulate_self_pow',
    'linear_dist',
    'cyclic'
]
```

```python
n = 100
m = 3
d = 2

n_neighs = [2, 4, 6, 8, 99]
```

```python
desc = 'regular'

if desc in desc_to_run:
    for n_neigh in n_neighs:
        G, T, L, obj = P.po.solve_opt(n=n, m=m, d=d, n_neigh=n_neigh, asrange=True)
        pl = P.pl.plot_sol(G, T, L, obj=obj, tit=None, paper=True, remove_labels=True)
        if save:
            fname = os.path.join(outdir, 's2_%s_%d.png' % (desc, n_neigh))
            pl.save(fname, scale_factor=5)
        pl.display()
```

```python
desc = 'exp_inhibit'

def h_exp_inverse(N, P, rate=1.5):
    return np.exp(-rate*(N @ P))

if desc in desc_to_run:
    for n_neigh in n_neighs:
        G, T, L, obj = P.po.solve_opt(n=n, m=m, d=d, n_neigh=n_neigh, h=h_exp_inverse, asrange=True)
        pl = P.pl.plot_sol(G, T, L, obj=obj, tit=None, paper=True, remove_labels=True)
        if save:
            fname = os.path.join(outdir, 's2_%s_%d.png' % (desc, n_neigh))
            pl.save(fname, scale_factor=5)
        pl.display()
```

```python
h_pow = 3#2.5

def h_stimulate_self_pow(N, P, h_pow=h_pow):
    return N @ np.power(P, h_pow)
    

def h_inhibit_self_pow(N, P, h_pow=h_pow):
    return (1 - (N @ np.power(P, h_pow)))

desc = 'stimulate_self_pow'

if desc in desc_to_run:
    for n_neigh in n_neighs:
        G, T, L, obj = P.po.solve_opt(n=n, m=m, d=d, n_neigh=n_neigh, h=h_stimulate_self_pow, asrange=True)
        pl = P.pl.plot_sol(G, T, L, obj=obj, tit=None, paper=True, remove_labels=True)
        if save:
            fname = os.path.join(outdir, 's2_%s_%d.png' % (desc, n_neigh))
            pl.save(fname, scale_factor=5)
        pl.display()
        
```

```python
desc = 'linear_dist'
if desc in desc_to_run:
    for n_neigh in n_neighs:
        G, T, L, obj = P.po.solve_opt(n=n, m=m, d=d, n_neigh=n_neigh, asrange=True, D_power=1)
        pl = P.pl.plot_sol(G, T, L, obj=obj, tit=None, paper=True, remove_labels=True)
        if save:
            fname = os.path.join(outdir, 's2_%s_%d.png' % (desc, n_neigh))
            pl.save(fname, scale_factor=5)
        pl.display()
```

```python
desc = 'cyclic'

def h_cyclic(N, P):
    return (1 - np.roll(N @ P, shift=1, axis=1))

if desc in desc_to_run:
    for n_neigh in n_neighs:
        G, T, L, obj = P.po.solve_opt(n=n, m=m, d=d, n_neigh=n_neigh, asrange=True, h=h_cyclic)
        pl = P.pl.plot_sol(G, T, L, obj=obj, tit=None, paper=True, remove_labels=True)
        if save:
            fname = os.path.join(outdir, 's2_%s_%d.png' % (desc, n_neigh))
            pl.save(fname, scale_factor=5)
        pl.display()
```
